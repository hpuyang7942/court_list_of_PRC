```python
def seperate_two(input_list):
    for i in range(len(input_list)):
        temp_list = input_list[i].getText().split(u'法院')
        middle_trial.append(temp_list[0])
        for j in range(1, len(temp_list)):
            primary_trial.append(temp_list[j])
        temp_list = []
```

```python
for i in range(len(url_list)):
    ## get the final url to use
    ## final_url = 'https://zh.wikipedia.org' + url_list[i]
    url = url_list[i]
    response = urllib.urlopen(url)
    data = response.read()
    soup = BeautifulSoup(data)
    
    response = urllib.urlopen(url)
    assert response.getcode() == 200
    
    if (i == 2):
        supreme = soup.find('div', attrs = {'class' : 'mw-parser-output'}).find('span', attrs = {'id' : '.E7.8E.B0.E5.88.B6'}).findParent().findNextSibling().find('li').find('a')
        middle = supreme.findParent().find('li')
        middle_N_primary = middle.findNextSiblings()
    elif (i == 15):
        supreme = soup.find('div', attrs = {'class' : 'mw-parser-output'}).find('div', attrs = {'id' : 'toc'}).findNextSibling().findNextSibling().find('a')
        supreme_trial.append(supreme.getText())
        middle_N_primary = supreme.findParent().findParent()
        for j in range(17):
            middle_N_primary = middle_N_primary.findNextSibling().findNextSibling()
            seperate_two([middle_N_primary])
        print 'Province ', i, ' has been recorded!'
        continue
    else:
        supreme = soup.find('div', attrs = {'class' : 'mw-parser-output'}).find('ul').find('li').find('a')
        middle = soup.find('div', attrs = {'class' : 'mw-parser-output'}).find('ul').find('li').find('ul').find('li')
        middle_N_primary = soup.find('div', attrs = {'class' : 'mw-parser-output'}).find('ul').find('li').find('ul').find('li').findNextSiblings()
    
    supreme_trial.append(supreme.getText())
    seperate_two([middle])
    seperate_two(middle_N_primary)
    print 'Province ', i, ' has been recorded!'
    
    if (i == 30):
        print 'Crawwling succed!'
```

```python
def is_chinese(uString):
    temp = ''
    for i in range(len(uString)):
        """判断一个unicode是否是汉字"""
        if uString[i] >= u'\u4e00' and uString[i] <= u'\u9fa5':
            temp = temp + (uString[i])
    return temp
```

```python
primary_trial[0][0] >= u'\u4e00' and primary_trial[0][0] <= u'\u9fa5'
```

```python
for i in range(len(primary_trial)):
    primary_trial[i] = is_chinese(primary_trial[i])
```
